## Compare claims by age groups

```python
# Define age ranges for the age distribution
df['Age cuts'] = pd.cut(df['Age'], 10)
df['Age cuts'].unique()
```

```python
# Define age groups manually
df['Age Groups'] = df['Age'].map(
    lambda x: '0+' if x < 10 else '10+' if x<20 else '20+' if x < 30 else '30+' if x < 40 else '40+' if x < 50
    else '50+' if x < 60 else '60+' if x < 70 else '70+' if x < 80 else '80+' if x < 90 else '90+' if x < 100 else '100+' if x < 110 else '110+')
```

```python
# Distribution in different age groups
df['Age Groups'].value_counts()
```

## Do older customers make more or fewer claims than younger customers?
- More than 50% of the claims are from people aged 30-39

```python
# Derive claim rejection rates for age groups
(df[df['Claim'] == 'No']['Age Groups'].value_counts() / df['Age Groups'].value_counts()).round(2)
```
